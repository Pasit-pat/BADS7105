```python
import pandas as pd
from scipy.spatial.distance import cosine
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head(6)
```

```python
#remove columns 1 out
data_ntime = data.drop('ประทับเวลา',1)
```

```python
#Create blank table 
data_ibs = pd.DataFrame(index=data_ntime.columns,columns=data_ntime.columns)

data_ibs.head(5)
```

```python
#Calculate cosine and input value in table
for i in range(0,len(data_ibs.columns)):
    for j in range(0,len(data_ibs.columns)):
        data_ibs.iloc[i,j]=1-cosine(data_ntime.iloc[:,i],data_ntime.iloc[:,j])
        
data_ibs.head(5) 
```

```python
#Display top 10 recommendation
data_neighbours = pd.DataFrame(index=data_ibs.columns,columns=range(1,11))

#Ascending top 10 recomendation
for i in range(0,len(data_ibs.columns)):
    data_neighbours.iloc[i,:10]=data_ibs.iloc[0:,i].sort_values(ascending=False)[:10].index
    
data_neighbours.head(5)
```
